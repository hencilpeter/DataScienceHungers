```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.sans-serif'] = "cmr10"
import random
```

```python
np.random.seed(54)
```

```python
def dist_function(x):
    f_x = np.sin(x)
    return f_x
```

```python
def gen_data(x):
    f_x = dist_function(x)
    range_value = 0.4
    # random.uniform - return floating point number between two numbers 
    y = f_x + np.random.uniform(-range_value, range_value, len(x))
    sigma = 1/12 * (2 * range_value)**2
    return y, f_x, sigma
```

```python
def get_rand_data(len_ratio, x, y, f_x):
    length_x = len(x)
    length_modified = int(np.round(len_ratio * length_x))
    index = random.sample(range(length_x), length_modified)
    index = np.sort(index)
    x1 = x[index]
    y1 = y[index]
    f_x1 = f_x[index]
    
    return x1, y1, f_x1
    
```

```python

x_orig = np.arange(0, 4 * np.pi, 0.2)
y_orig, f_x_orig, sigma = gen_data(x_orig)
p_order = [1,3,5,7,9,11]
num_data = 2000
```

```python
f, ax = plt.subplots(1, len(p_order), figsize=(15,1.5), dpi=120, facecolor='w', edgecolor='k')
temp = np.arange(0.1, 4 * np.pi, 0.2)
polynomial_functions = []
bias_values = []
variance_values = []

x, y, f_x = get_rand_data(0.1, x_orig, y_orig, f_x_orig)

for i, p_value in enumerate(p_order):
    exp_f_x = np.zeros(len(temp))
    exp_f_x_plot = np.zeros(len(x_orig))
    variance = np.zeros(len(temp))
      
    for j in range(num_data):
        x, y, f_x = get_rand_data(0.7, x_orig, y_orig, f_x_orig)
        #polyfit - fit the polynomial and return coefficients 
        # poly1d - return the polynomial function. later its easy to apply natural operation. 
        poly_function = np.poly1d(np.polyfit(x, y, p_value))
        polynomial_functions.append(poly_function)
        exp_f_x = exp_f_x + poly_function(temp)
        exp_f_x_plot = exp_f_x_plot + poly_function(x_orig)
        
    exp_f_x = exp_f_x/num_data
    exp_f_x_plot = exp_f_x_plot / num_data
    
    #draw the current polynomial curve/fit 
    ax[i].plot(x, y, 'o', x_orig, exp_f_x_plot, '-')
    plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[])
    ax[i].set_xlabel('P='+str(p_order[i]))
    
    #calculating bias
    bias = np.linalg.norm(exp_f_x - dist_function(temp))
    bias_values.append(bias)


    #calculating variance 
    for j in range(num_data):
        poly_function_current = polynomial_functions.pop(0)
        variance = variance + np.square( poly_function_current(temp) - exp_f_x )
    variance = variance / num_data
    variance = np.linalg.norm(variance)
    variance_values.append(variance)
    
    
print('Bias : ', bias_values)
print('Variance  : ', variance_values)
    
    
    
#########

fig, ax1 = plt.subplots(dpi=200)

#plot variance
color = 'tab:orange'
ax1.set_xlabel('Polynomial order')
ax1.set_ylabel('Variance', color=color)
ax1.plot(p_order, variance_values, color=color, label = 'Variance')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

#plot bias
color = 'tab:blue'
ax2.set_ylabel('Bias', color=color)
ax2.plot(p_order, bias_values, color=color,  label = 'Bias')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()

#plot total error 
error = np.square(bias_values)+variance_values+sigma*len(temp)
min_ind = np.argmin(error)
plt.plot(p_order, error, 'k--', label = 'error')
plt.plot(p_order[min_ind], error[min_ind], 'ro', label= 'Minimum Error')
leg = ax2.legend(loc='upper right')
leg = ax1.legend(loc='upper left')
plt.show()
```
